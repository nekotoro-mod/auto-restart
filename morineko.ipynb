```python
# ==============================
# Minecraft Paper サーバースクリプト (Google Colab + Drive)
# ==============================

import os
import urllib.request
from google.colab import drive

# ==============================
# Google Drive マウント
# ==============================
print("Google Drive をマウント中...")
drive.mount('/content/drive2', force_remount=True)

# サーバーデータ保存先
SERVER_DIR = "/content/drive2/MyDrive/minecraft_server"

# ディレクトリ作成
os.makedirs(SERVER_DIR, exist_ok=True)

# ==============================
# JDK 21 のインストール
# ==============================
print("JDK 21 をダウンロード＆インストール中...")
!apt-get update -y
!apt-get install -y openjdk-21-jdk
JAVA_PATH = "/usr/lib/jvm/java-21-openjdk-amd64/bin/java"

# ==============================
# PaperMC のダウンロード
# ==============================
PAPER_JAR = os.path.join(SERVER_DIR, "paper.jar")

if not os.path.exists(PAPER_JAR):
    print("PaperMC をダウンロード中...")
    url = "https://api.papermc.io/v2/projects/paper/versions/1.21/builds/126/downloads/paper-1.21-126.jar"
    urllib.request.urlretrieve(url, PAPER_JAR)
else:
    print("既存の PaperMC を使用します。")

# ==============================
# EULA 同意
# ==============================
eula_path = os.path.join(SERVER_DIR, "eula.txt")
if not os.path.exists(eula_path):
    with open(eula_path, "w") as f:
        f.write("eula=true\n")

# ==============================
# サーバー起動
# ==============================
print("サーバーを起動中...")

# 進捗表示（例: 0〜100%）
import time
for i in range(0, 101, 10):
    print(f"起動準備中... {i}%")
    time.sleep(0.2)

# 実行
%cd $SERVER_DIR
!{JAVA_PATH} -Xmx2G -Xms1G -jar paper.jar nogui
```
